```python
class Human:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      return f"{self.name}は大人です"
    else:
      return f"{self.name}は大人ではありません"


human1 = Human("侍太郎",22)
human2 = Human("侍花子",20)
human3 = Human("侍一郎",15)

human_list = [human1,human2,human3]

for human in human_list:
  print(human.check_adult())

```
